```python
import numpy as np
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
import xml.etree.ElementTree as ET 
```

```python
#BasePath = "../DNA_DATA/SGA1/"
```

```python
def get_data(BasePath):
    tree_max = ET.parse(f'{BasePath}/ScoreMax.xml')
    tree_min = ET.parse(f'{BasePath}/ScoreMin.xml') 
    tree_mean = ET.parse(f'{BasePath}/ScoreMean.xml') 
    tree_std = ET.parse(f'{BasePath}/ScoreStd.xml') 
 
    # XMLを取得
    root_max = tree_max.getroot()
    root_min = tree_min.getroot()
    root_mean = tree_mean.getroot()
    root_std = tree_std.getroot()
    
    scoreMax = []
    scoreMin = []
    scoreMean = []
    scoreStd = []
    
    for i in range(len(root_max)):
        scoreMax.append(float(root_max[i].text) )
        scoreMin.append(float(root_min[i].text) )
        scoreMean.append(float(root_mean[i].text) )
        scoreStd.append(float(root_std[i].text) )
    scoredf = pd.DataFrame()
    scoredf["maxScore"] = scoreMax
    scoredf["minScore"] = scoreMin
    scoredf["meanScore"] = scoreMean
    scoredf["stdScore"] = scoreStd
    
    data_dict = {"name":[]}
    dna_path = glob(f"{BasePath}/DNA*")
    path = dna_path[0]
    name = path.split('DNA_')[-1]
    tree = ET.parse(path)
    root = tree.getroot()
    data_dict["name"].append(name)
    for c in root:
        data_dict[c.tag] = []
        data_dict[c.tag].append( int(c.text) )
    for path in dna_path[1:]:
        tree = ET.parse(path)
        root = tree.getroot()
        name = path.split('DNA_')[-1]
        data_dict["name"].append(name)
        for c in root:
            data_dict[c.tag].append( int(c.text) )
    dnadf = pd.DataFrame(data_dict)
    return scoredf, dnadf
```

```python
alg_names = ["SGA", "IGS", "SS", "CHC", "ER", "MGG"]
score_df_sga, dna_df_sga = get_data(f"../DNA_DATA/{alg_names[0]}_EXP1/")
score_df_igs, dna_df_igs = get_data(f"../DNA_DATA/{alg_names[1]}_EXP1/")
score_df_ss, dna_df_ss = get_data(f"../DNA_DATA/{alg_names[2]}_EXP1/")
score_df_chc, dna_df_chc = get_data(f"../DNA_DATA/{alg_names[3]}_EXP1/")
score_df_er, dna_df_er = get_data(f"../DNA_DATA/{alg_names[4]}_EXP1/")
score_df_mgg, dna_df_mgg = get_data(f"../DNA_DATA/{alg_names[5]}_EXP1/")
```
